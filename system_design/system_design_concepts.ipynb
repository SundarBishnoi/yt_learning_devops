SD - Art of being able to build large scale system.

Skill, we shall develop here is problem solving. PS is if i describe a scenario to you, and pause challenge to you, how will you architect in this scenario. are you able to list down all corner cases and design that works for all corner cases. Its not about knowledge, its being able to interpret scenarios, list down corner cases and build the right design for what you're building for.
