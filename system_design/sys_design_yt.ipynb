Restaurant Analogy, you have a pizza shop!

Pizza Parlor - 1 Chef 
As 1 chef cann't handle all the orders coming in. 

How do you handle more orders?

1. Optimize processes and increase throughput using the same resource => Vertical Scaling. 
2. Preapring before hand at non-peak hours. => Pre=processing and cron jobs

Resilience - recovering from failures quickly. 
1 chef is doing all - single point of failure, so hire a backup chef. 

3. Keep backups and avoid single points of failure => Backups => Master/Slave Architecture

if business keep growing, hire more chefs, so now you have 10 chefs, 3 backups

4. Hire more resources => Horizontal scaling(Buying more machines of similar types to get more work done)

5. Microservice architecture => you have all your responsibilites well defined, there is nothing outside your business case that you handle.

6. Distributed System(Fault tolerant) with partitioning(Quicker response time).

7. Central place which just routes requests, based on very clear parameter(how much time does it take for customer to get the pizza)
Route requests in smart way is called load balancer. 

can it be future-proof? system is now fault tolerant? how do we make it flexible to change. 

8. Seperating out concerns so that you can handle seperate systems more efficiently is decoupling. 

9. Logging and metrics calculations. 

Final and most important point is to keep your system extensible.
