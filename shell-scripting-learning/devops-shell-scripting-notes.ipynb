```text
Shell scripting always starts with shebang. special indentation => #!/bin/bash OR #!/bin/sh OR #!/bin/ksh

why would everyone writes shebang inside shell script?
    -   These are different executable of your linux, depending on kind of linux machine - bash, sh, dash, ksh
        and they will execute your script.. by writing at top, you are telling, i want to use this executable to run my script..
    -   bash is most popularly used. 
```


```text
Difference b/w #!/bin/sh and #!/bin/bash
    - They are same as #!/bin/sh will redirect to #!/bin/bash using linking in linux. but recently few distributions started redirecting to #!/bin/dash so always use #!/bin/bash for safer approach.

```
