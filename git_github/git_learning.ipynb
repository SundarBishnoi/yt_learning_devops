# Strategy used to apply hot fix changes in git

This method isolates the urgent fix from ongoing development, tests it, and applies it to both production and development environments.

step-by-step workflow

- create a dedicated hotfix branch: 
    Always branch directly out of your production branch (usually main or master). This ensures you only fix the bug without accidentally pulling in unfinished features from your development branch

    ```python
    git checkout main
    git pull origin main
    git checkout -b hotfix/bug-description
    ```
- apply and test the fix:
    Make the necessary code changes on this isolated branch. Run your local tests to verify the fix works and does not break existing functionality.

    ```python
    git add .
    git commit -m "fix: resolve critical login page crash"
    ```
-  merge into Production:
    Once verified, merge the hotfix branch back into your production branch and deploy it immediately.
   ```python
    git checkout main
    git merge hotfix/bug-description
    git push origin main
   ```
-  Tag the Release:
    Create a Git tag on the production branch to mark the new patch version (e.g., changing version v1.2.0 to v1.2.1)
```python
    git tag -a v1.2.1 -m "Hotfix release for login crash"
    git push origin v1.2.1
```
-  Sync with Development:
   To ensure the bug doesn't reappear in future releases, merge the hotfix branch back into your development branch (usually develop or next).
```python
    git checkout develop
    git merge hotfix/bug-description
    git push origin develop
```
-  Delete the Hotfix Branch:
    Clean up your local and remote repositories by deleting the temporary hotfix branch.
```python
    git branch -d hotfix/bug-description
    git push origin --delete hotfix/bug-description
```

# 🚀 Enterprise & Open-Source Git Master Workflow

## Phase 1: One-Time Workspace Setup
1. Fork the official repository on GitHub to your account.
2. Clone your personal fork to your laptop:
   git clone https://github.com
3. Link the official repository as your "upstream" track:
   git remote add upstream https://github.com

## Phase 2: The Daily Morning Sync
Always start your day by aligning your local machine and your cloud fork with the world:
1. git checkout main
2. git pull upstream main      # Grabs the community's latest work
3. git push origin main        # Updates your personal GitHub fork website

## Phase 3: Isolated Feature Sandbox Development
1. git checkout -b feature-devops-milestone
2. [Write code, test configurations, modify Dockerfiles/YAMLs]

## Phase 4: Stage, Commit, & Sync Check
When your code works perfectly local, freeze your checkpoint:
1. git add .
2. git commit -m "feat: added optimized multi-stage build to Dockerfile"
3. git pull origin feature-devops-milestone  # Check for branch-specific remote changes
4. git push origin feature-devops-milestone  # Push to YOUR cloud fork branch

## Phase 5: The Pull Request & Hands-Off Merge
1. Go to the original repository on GitHub and click "Open Pull Request".
2. Link your `feature-devops-milestone` branch for peer review.
3. 🛑 STOP HERE. Wait for the project maintainers to approve and click MERGE.

## Phase 6: Post-Merge Housecleaning
Once the maintainers merge your PR on the GitHub website, clean your local machine:
1. git checkout main
2. git pull upstream main      # Pulls your own newly approved code back down to your machine
3. git branch -d feature-devops-milestone  # Safely delete your temporary sandbox branch


# 🗺️ Complete Developer Git Framework

## 1. Initial Project Workspace Setup
1. Fork target repository on GitHub to your account.
2. Clone your personal fork locally:
   git clone https://github.com
3. Add the official project as your 'upstream' source link:
   git remote add upstream https://github.com

## 2. Daily Morning Sync Loop
1. git checkout main
2. git pull upstream main      # Fetch & merge latest community code
3. git push origin main        # Mirror clean updates to your GitHub fork UI

## 3. Standard Feature Delivery Flow
1. git checkout -b feature-my-devops-update
2. [Develop requirements & test configurations locally]
3. git add .
4. git commit -m "feat: implemented core automation module"
5. git push origin feature-my-devops-update
6. Open a PR on GitHub from your fork branch to upstream main.
7. 🛑 STOP. Wait for peer review and maintainer approval to merge.

## 4. Emergency Outage & Hotfix Interruption Flow
Scenario: You have uncommitted, untested changes on your feature branch when a critical outage strikes.

### Step 4A: Freeze & Pivot
1. git stash                       # Safely hide unfinished feature work
2. git checkout main               # Switch to the production baseline
3. git pull upstream main          # Get latest state
4. git push origin main            # Update your fork
5. git checkout -b hotfix-bug      # Create dedicated triage branch

### Step 4B: Patch & Deploy
1. [Apply emergency code fix & test thoroughly]
2. git add .
3. git commit -m "fix: patched critical production outage bug"
4. git push origin hotfix-bug
5. Raise emergency PR to upstream main. Maintainers review, merge, and cut release branch to deploy.

### Step 4C: Clean Up & Sync After Deployment
1. git checkout main
2. git pull upstream main                 # Pull your hotfix back down into main
3. git push origin main                   # Push it up to your fork's main
4. git branch -d hotfix-bug               # Delete local hotfix branch
5. git push origin --delete hotfix-bug    # Delete remote hotfix branch from your cloud fork

## 5. Resuming Feature Work & Resolving Stash Conflicts
Scenario: The hotfix modified the exact same files you stashed earlier.

1. git checkout feature-my-devops-update   # Get back to your feature sandbox
2. git merge main                         # Pull hotfix changes into feature branch FIRST
3. git stash pop                          # Bring back hidden work (Triggers conflict warning!)

### Conflict Resolution Steps (If files conflict):
1. run 'git status' -> Conflicted files appear in RED under "Unmerged paths".
2. Open conflicted files in your editor and manually clean the <<<<<<, ======, >>>>>> markers.
3. git add $filename                      # Tell Git the file is clean and resolved
4. git stash drop                         # Manually delete the stash backup copy

*Aborting Option:* If manual cleanup gets messy, run `git reset --hard HEAD` to completely wipe the popped changes and reset back to step 1 without losing your stash backup.


# Managing secrets (like API keys, database passwords, and SSH keys)

The golden rule of Git is: Never commit raw secrets to your repository. Once a secret is pushed to GitHub, it is considered compromised, even if the repository is private.Here is the industry-standard approach to handling secrets locally, in Git, and within your CI/CD pipelines.


1. Local Development: Use .env and .gitignoreWhen writing code locally, keep your configuration and secrets separate from your source code using environment variables.
- The Code: Use libraries like python-dotenv (Python) or dotenv (Node.js) to read variables from a local file.
- The .env File: Create a file named .env in your project root to hold your keys:
    DB_PASSWORD=super_secret_password_123
    API_KEY=AIzaSyD-ExampleKey

2. The .gitignore File (Crucial): Immediately add .env to your .gitignore file so Git completely ignores it:
-  Ignore local secrets
    - .env

3. The .env.example File (Best Practice): Create and commit a dummy file to show other developers what keys they need to set up without exposing the real values:
- DB_PASSWORD=your_database_password_here
- API_KEY=your_api_key_here


----------------------

2. In the Cloud: Use Repository Secrets (GitHub Actions) 
    When your CI/CD pipeline runs or your project needs deployment, you inject the secrets securely through your repository provider instead of putting them in your files.Go to your GitHub repository → Settings → Secrets and variables → Actions. Click New repository secret.Name your secret (e.g., PROD_DB_PASSWORD) and paste the value.Reference it securely in your GitHub Actions YAML pipeline workflow using standard brackets:

```python
    - name: Run Database Migration
      env:
        DB_PASSWORD: ${{ secrets.PROD_DB_PASSWORD }}
      run: python migrate.py
```

3. Production Environments: Secret Managers 
        For enterprise applications (like the systems used at OneTick or Accion Labs), production secrets are stored in dedicated, encrypted vaults that apps pull from at runtime:HashiCorp Vault: The industry standard for platform-agnostic secret management.AWS Secrets Manager / Azure Key Vault / GCP Secret Manager: Cloud-native tools that automatically rotate and manage keys safely.
    


**Emergency Triage**: "I accidentally pushed a secret to GitHub!"

If you realize you committed a raw secret to a remote repository, do not just delete the file and make a new commit. The secret remains visible in your Git commit history forever. Take these steps immediately:

1. Revoke the Secret Immediately: Change the password or rotate the API key on the provider's platform (e.g., AWS, Google Cloud). This makes the exposed key useless.
2. Purge Git History: Use an open-source tool like bfg-repo-cleaner or run git-filter-repo to completely strip the string from your entire repository history before force-pushing back to your fork.
