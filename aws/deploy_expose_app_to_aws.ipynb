IAM -> identity and access management. 
you have lot of different people working under you. you are the head. so as a head, you definetly would have access to root account. but you don't want that each and every team has access to certain credential/service. so basically, we assign roles to all the different people in a particular teams which are working under you. for that, aws provides services, which is called IAM through which we actually give roles to people. Basically, you would typically have role based access control system. using this, you can restrict the services, and fine tune read, write, execute accesses on aws..
Basically, there are 3 ways, you can give permissions. 
1. You can create group permission. 
2. Copy permissions
3. Attach policies directly.. which can be modified/updated later as well. 

ec2 - elastic compute cloud, basically used to setup remote server. maybe, if you have old laptop, you can get compute resources on these cloud providers. Even in organization, you will use these resource as these cloud provider can remove your maintenance overhead which you get on physical server with hypervisors as with every security vulnerability, you will need to keep upgrading them. Another advantage is cost can go down if you efficiently use them. 
while creating ec2 instance. you have option to go into particualr AZ. primary reason it helps with latency in real. for indian user, good practice is to create in mumbai location as its not far from your laptop while doing ssh. AWS keeps its datacenter in multiple locations those are called regions.
