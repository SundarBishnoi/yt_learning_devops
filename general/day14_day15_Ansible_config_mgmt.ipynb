Why ansible is used very widly in the field of config management?
-   config mgmt/infra mgmt
-   config mgmt - it is the way how the devops engineer manage the configurations of server or config of infrastructure.

config mgmt is a way, how devops engineer manage the configuration of the servers or infrastructure.

DevOps only came to evolution when all of these things merged together. lets say you're sys admin and your company was using on-premises server. i.e. 100s of server(your own server hosted on data center). Now to improve the performance of distribution, or due to security patches, you will have to upgrade to next version. 

1. Upgrade
-   Basically, To improve the performance of the distribution. 
2. Security Patches
-   You have to do upgrade due to vulerability patches to next version. 
3. Installation
-   There might be default installations that your organization might require like git

Scalaing this to 100s/1000s/10000s Server using shell script is techincally very difficult. 
And when all of this is migrated to cloud, no of servers that were getting created has increased by 10X and size got reduced by 10x.
So now the problem is bigger. we have to keep writing script per the platform requirement(shell for linux, powershell for windows so on) so concept of config management was discovered. 

config mgmt will aim to solve the problem of managing the config of multiple servers(Due to advancement of cloud and microservices). 
Inside config mgmt, there are popular tools like, puppet, Ansible, Chef, Salt. Eventual winner is Ansible(Most Devops found this better)/Goto tool for config mgmt.
Ansible is constantly developed(updating modules) by people at Redhat.

Why Ansible?

|------|-------|

||In this dynamic environment, you can always scale down or scale up your server at any given time. lets say, there is christmis, you're expecting company's load to go up. on specific, IRCTC is going to scale up their servers. And when there is less traffic, you can always scale down.|
||Ansible also has dynamic inventory, ansible will auto detect, now you don't even have to go and update your inventory file.. previously, you had to go and update your inventory file. Lets say, you're creating 100s new ec2 instance in AWS account, Ansible is capable by making some settings/by having ini file/ and making some changes to this file, you can ask ansible, whenever new ec2 instance is created on my aws account in a specfic region or specific availability zones, directly consider that a server that has to be managed by you. |
||Support with windows & linux is very easy with both|
|Need to learn new puppet language|Very simple in writting due to simple yaml manifest which gives fundamental advantage. Language that is used in ansible is yaml.|
||Ansible is written in python, so you can write your own module and anybody in the world can use that like open source tool. Using ansible galaxy, you can create module and share it with other devops in different organization|

```python
18 Most Asked Interview Questions

1. What is configuration management?
Its a process of managing your software system or hardware. Lets say, in your organization. you have 100s of servers(be it cloud or on-premises). you want to manage installation of couple of packges or upgradation of couple of software. and you have to do some mandatory installation. be it CVE or anything. for which you use config management. 

why do you use that? - because if you want to do it manually on all of these server or using some scripts. it will take a lot of time. Each of these servers maybe different OS/or of different flavour of similar OS because of which you might have to end up writing multiple scripts. or you might have to so some manual patching. so to avoid this, you can use a single tool which supports the config managment like ansible/puppet/chef/salt. 

2. Do you think ansible is better than other config tool? if yes, then why?
Ansible is better than config mgmt tool. its better in achieving a specific usecase. Ansible is agentless. you dont need to configure any agents which other tool requires like master slave archi. only prereq. is that you need to have password less authentication b/w your host that has ansible installed and all the other server that you want to configure with ansible. 
Anisble is written in python and backed up by good community. 

3. can you write ansible playbook to install httpd service and get it running?
install the httd service
start the httpd service
check if needed to check if firewall is open. 

4. How ansible helped your organization?
based on exposure.. previosly it was manual step. with anisble automation, saved time to upgrade. 

5. What is ansible dynamic inventory?
let's assume you have VM on AWS and now you have configured ansible to make some configuration on that VM. lets assume that today you have 50 VMs and ansible is very effective in configuring those 50 VMs. i.e. you're doing mysql db installation on all of these VMs. Later point, servers are 60. now you dont add them manually to inventory so for this. ansible supprts dynamic inventory. Ansible keeps looking at your EC2/AWS account for what are newly configd VMs. ansible would auto configure those VMs using the concept of dynamic inventory. 

6. What is ansible tower and have you used it? if yes, why?
not everyone works on enterprise model/version of ansible where it provides you a fancy graphical UI and it also extends some additional ansible capabilites. It has good user management. you can configure with your IAM if you're using ansible tower on AWS. Also you get support if there is any issue. if not worked, then you can say we are using only open source version of ansible. 

7. How do you manage RBAC of users on Ansible towers?
Here, you are asked about role based access control. RBAC is used to managing the access on ansible tower. you can also configure with some external identity provider/SSO like ldap or key clock or any other external identity provider. you can also integrate them with ansible tower. or if you are on aws, you can also configure with IAM. 

8. What is ansible galaxy command and what is it used for?
ansible galaxy command is used to bootstrap. lets assume you want to create whole structure, instead of creating whole strucutre of project on your own, ansible can do that.

9. can you explain me structure of ansible playbook using roles?
ansible-galaxy command which creates the structure of your playbook and it creates roles, it creates all your default, handlers everything. 
basically with roles you will have default, handler, templates, README.md, metadata, tasks folders. you can also create them manually. you can descrive what each and every folder does. 

10. What are handlers in ansible and why are they used?
handler are basically similar to the normal task but they only run when task contains notify. they basically only runs during a special notification. Lets say, your task is to start nginx but for some error, it could n't start then you can notify handler, run specific task to send a email.

11. I would like to run a specific set of tasks only on windows vms and not linux vms, is it possible?
Ansible supports tags and you can find out the environment variable and you can use the conditional basis running of tasks in ansible to achieve this use case. 

12. Does ansible support parallel execution of tasks?
lets say if there are 5 tasks, ansible does execute task number 1 on all the configured servers. Lets assume, ansible wants to run on 10 VMs. It takes task1 and it runs parallely on task1 on all of these VMs. and then it runs task by task in a serial way that means once task1 is done execution parallelly on all of the hosts then it moves to task 2. if you want to run even task in parallel way, then you need to have multiple instances of ansible. 

13. What is the protocol that ansible uses to connect to windows VMs
ansible uses win rm to connect to windows and it uses SSH to connect linux 

14. Can you place them in the order of precedence?
you have multiple places to store variables in ansible. Order of precedence => group_vars, role_vars, extra_vars

15. How do you handle secrets in ansible?
you can configure vaults. hashicorp has something called as vault so you can store the ansible secrets in the vault. 

16. Can we use Ansible for IaC? Infrastrucutre as Code? If yes, can you compare it with any other IaC tools like Terraform?
predominately ansible is configuration management tool but it also supports IaC. Ansible can also achieve some of the tasks that TF can do. Like it can create EC2 instacne on aws. but mostly its used for configuration  management.

17. Can you talk about a Ansible playbook that you wrote and how it helped your company?
Talk about Something that you eliminated manual efforts and saved time. 

18. What do you think Ansible can improve?
Currently ansible does not provide verbosity of the tasks. Ansible can improve in that aspect, it can provide verbosity on the task basis. 
Ansible windows support can be improved. Support for linux is much better than linux. 
Ansible can support multiple IDEs but there are not many IDE that can do auto-correction/suggestion. 



```
