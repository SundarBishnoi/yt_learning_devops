```python
Infrastructure as Code - its concept using which you can autoamte your infrastructure. it can be anything like writing CFT, azure RM. 
API as code - 

Theory  -   Problem, usecase, solution. 

lets say flipkarts wants to create CPU, RAM or any other hardware resources to deploy their 300+ applications.
They can create on any of the cloud platform or on premises. 
As a devops engineer, after evaluating all of the options, you decided to move with AWS. 
you started to create ec2 instance, s3 bucket, or it can be RDS database. all of these resources are required for your organization. As a good devops engineer, you have started to automate the entire process on AWS. Like instead of manually creating through console, you have decided to use AWS CLI or AWS CFT. 
you have used AWS CFT and have automated the infrastructure on AWS cloud platform. now based on requirement, develpoer requirement, using cft automation scripts, you can provide the required infra component in no time. 
But now, due to some reason(AWS cost or support provided, you're not happy with it), management decided to shift from aws to azure. now these CFT scripts are of no use when required as they only belong to AWS cloud provider. now you've migrate this 100s of scripts into azure resource manager. And again you have automated and converted the entire process. and now even support with Azure is not good or not happy with costing. and this time, you want to move to on-premises. now you want to maintain your own infra.(by purchasing server and putting into data center) and on top of it, you use openstack to create your own cloud. you created the entire data center, you installed openstack. and to fullfill the infra requirement for developer team(i.e. 10 VMs), you again migrate these scripts that were firstly on AWS CFT, then with Azure RM, now again you have to migrate them to heat template(its way how openstack automate the process). now a days, companies are operating in hybrid cloud model. and any organization depending on requirements, they can host, part of their infrastructure on AWS, and part of their infrastrucutre on Azure, and on-prem too. lets say AWS is giving good supprt with respect to storage services. and for devops related service(build related) and project management related service, you want to use Azure DevOps. now you need to learn both CFT and RM and heat map for onprem cloud services where latency/compliance/security/auditting are mandatory.
this can extend beyond. so a devops engineer, which tool, will you learn. and to solve this, people had compe with Terraform which addresses this problem of learnign too many tools. 
TF is a tool developed by Hashicorp. which solves the problem of learning too many tools or execution of too many tools.
Terraform ------>>>>>> API as Code ------>>>>>>> 

Create TF Script. TF will take care of automating the resources whether its with AWS, or Azure or GCP. TF will take care of automating the resources on cloud provider. 
So even if you move from 1 cloud provider to another cloud provider, changes required in TF scripts will be less compared to sticking with CFT or RM as you just have to update provider details, modules or other minimal changes. And just looking at TF documentation, you can write any automation for any infrastructure. So, this concept is called as API as code. API as code, is a concept using which you can automate any cloud provider using their APIs. TF will talk to API of AWS/azure/openstack/gcp. So once you write the TF templating/script files, TF will convert them depending upon provider details that you have provided. So in 1 of file, called provider.tf, you say that the provider is AWS. so once you write the TF module, TF will convert that script into AWS readable API or request which aws can understand. and it executes the action, and gives you the results back. 
This is how TF is one tool which can automate the resources on any cloud provider. 
API - it's nothing but application interface so using API, you can talk to any application and get the response back. lets you want to talk to API, usual process is you open your laptop, and on your laptop, you will open your favorite browser, you will type www.google.com which sends the requests to google. you get the browser page, and you can search the information related and this is basically a user interface. So this is a way, how you do manually things. but whenever you want to automate things, or whenever you want to do things programmatically, this way is not suitable. Lets say, you want to get some information on google in programmatical manner. so instead of logging into laptop, opening the browser, and getting the information from Google, what you want to do is you want to run a script and this script will talk to google. and it executes the action and sends you the result back. To perform this kind of things, what developer have come, whether it is google, ArgoCD or any applcation, they have come up with concept of API. Using API, you can programatically talk to any application. So whether, its google, Github, Or any other application. Basically, all application/github will expose their API.
eg, Github tells that instead of manually looging into github, authenticating with me. and opening your browser and going through flows, you can open your terminal and use a curl request or you can use http get request. and talk to my API and get whatever information that you want, i'll send you the result. 
Instead of you talking to these provider API using python or shell, TF is mediator which will actively look at these API and TF has written their own modules. 

Lets say, you want to create EC2 instance on AWS. TF said that instead of you directly making API call with AWS. you can write some TF files. and in this TF file(you can look at documentation). TF has a module called ec2 using which you can write a bunch of lines like you write in English. and once user submits this requests. in TF file, once you do this. TF will receive this request, and it will take your input and it will convert your input into API call in backend. and it gets the request back and again TF will send you the request back to you. so as a user, you're not directly talking to APIs of AWS, rather you're writing TF scripts and TF is converting your requests into API calls. this is the concept of TF and this is how, TF utilizes the concept called API as code. 

```
