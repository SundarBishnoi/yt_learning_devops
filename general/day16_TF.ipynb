```python
In a ideal world, this is how TF looks like. Devops will write the TF script, they will store the tf script in git repository. and there are users who wants to use this tf script. as devops engineer, i did n't grant the access to aws infra. apart from devops, nobody logins to aws for some reason. no body has access to create resources on aws. if they want to create resources on aws, they should use the jenkin pipeline that devops created which will watch for tf files in github. it will pull the tf files from github. and it will execute the resources on aws. this person using output.tf will get the information of your resources created on aws. statefile will be stored in amazon s3 if you're using aws. once you store statefile locally, you will run into so many issues. even if you put that in github. but if you put in centralised location like S3, anybdy who is executing your tf script, tf will automatically update the statefile in centralised location/s3. one more challenge it addresses, there are 2 people who're executing this tf script parallelly so there are 100 people, who are using this tf script.. and tf shall not allow it. as they will be giving conflicting inforamtion to aws. and along with amazon s3, you will also integrate your remote backend with dynamo db. dynamo db is basically used for locking the terraform statefile. once it is locked, tf will say tha statefile is locked, some other user is actually using the tf script or performing the infra config on your aws resources so wait until, config is done, once the statefile in s3 are unlocked, then next user is allowed to execute the tf scripts. ideal config look like always put your tf config file in github but your tf statefile should go into remote backend(remote storage services(s3) and intergrate them with proper locking solutions, in case of aws, locking solution is dynamo db and statefile will not allow parallel execution of tf scritps)

isolate and organize statefile, as in segregate the tf file by environment like uat,qa,prod and you can reduce or implement the disaster recovery so in such cases. 
even if somebody is messing up with statefile accidently or for any reason. you can reduce the blast radius, in such cases, alway have statefile isolated. 

How to do remote backend?
only change needed, is we provide the backend details in main.tf of remote_state folder. in the existing tf config, you add the s3 bucket and dynamodb table. and these are 2 things, you do additionally from local -> main.tf
you can simply do in your existing config, you can directlly add these details as part of backend. once you create s3 bucket and dynamo db, add the details of backend. Apart from your region details that we are providing, we just add the backend s3 details. and your tf will configd. 
your tf will be configured with remote storage i.e. s3 bucket. 

Modules
-   Existing modules..
-   Write your own module..
whenever you have reusable code, lets say you have code for implementing ec2 instance plus ALB. you think, this code is used quite common or even if we take the same example, 
you might be doing same thing in dev/staging/prod. so this becomes reusable code because you can use this code across multiple teams or multiple environments. lets say its put in repo. you will start using that repo as module inside other tf scripts. you will reference these things as module in tf file. once you reference them as module in terraforms. before execution of these scripts, terraform will start executing the modules so whenever you're trying to run same example or trying to setup terraform, the users can worry about only writting the terraform files, and if they just start using your example as module, by deault. they will ahve tf configd with s3 remote backend and dynamodb enabled so lof of efforts has reduced. 

module is a way of writting reusable components whether it is terraform or jenkins or any other such thing. 


Problems with terraform
-   State file is single source of truth. Statefile is heart of Terraform.
-   Manual changes to the cloud provider cann't be identified and auto-corrected. its not bidirectional. with manual changes on cloud provider, next time tf will not be able to update as there is change in state on what it is with cloud provider. so now you have to manually update your statefile using terraform refresh command. 
-   Not a GitOps friendly tool, Don't play well with Flux or ArgoCD. (In gitops, git is single source of truth, but with manually changes terraform is not automatically correcting git files)
-   Can become very complex and difficult to manage.
-   Trying to positon as a configuration management tool as well.


Terraform interview questions
-   Explain me scenario where you had faced problem with terraform. 
-   what are modules in terraform. 
-   tell me your terraform setup. 
-   are you using statefile locally or in git repository(we use tf state as remote backend with as s3 and dynamo db for locking to avoid parallel execution)
-   what are the good practices with tf

```
