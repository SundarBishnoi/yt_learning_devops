Daily Git Workflow

Step 1: Start Fresh (Pull Remote Work)
Before writing any new lines of code, make sure your workspace is identical to your GitHub cloud records 

```python
# 1. Switch back to your core production tracking branch
git checkout main

# 2. Grab any remote adjustments or modifications
git pull origin main
```

Step 2: Safe Experimentation (The Feature Branch)
Never run model configurations or data analysis adjustments directly inside main [⚡]. Create an isolated sandbox branch for the day's specific milestone [⚡]:
```python
# 1. Create and switch to a targeted branch (e.g., 'data-cleaning')
git checkout -b feature-data-cleaning
```

Now open your Jupyter Notebook, evaluate your cells, and write code freely [⚡]. If you break a cell layout or corrupt a configuration state, your main line stays completely unaffected and safe [⚡].


Step 3: Save and Sync Up (Stage, Commit, & Push)Once your training modules print successful metrics, freeze that checkpoint securely [⚡]:

```python
# 1. Gather all modified notebook JSON layouts and updates
git add .

# 2. Label your progress log with a clear summary
git commit -m "completed outlier removal pipeline"

# 3. Push this specific feature up to your GitHub cloud panel
git push origin feature-data-cleaning
```

Step 4: The Final Review (Merge into Main)Once you are confident the new updates run flawlessly, weave them back into your main branch tree [⚡]:

```python
# 1. Jump back to your master production track
git checkout main

# 2. Merge your successful additions in cleanly
git merge feature-data-cleaning

# 3. Sync the updated main line back up to GitHub
git push origin main

# 4. Safe housecleaning: delete the branch wrapper when finished
git branch -d feature-data-cleaning
```
